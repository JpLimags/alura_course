```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df_tmdb = pd.read_csv(r"C:\Users\Jp Lima\Desktop\alura_course\datasets\tmdb_5000_movies.csv")
df_tmdb.head()
```

```python
total_other_movier_per_language = df_tmdb.query("original_language != 'en'")["original_language"].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="original_language", data=df_tmdb.query("original_language != 'en'"), order=total_other_movier_per_language.index,
              palette="mako", hue_order=total_other_movier_per_language)
plt.title("Distribuição dos filmes exceto inglês")
```
