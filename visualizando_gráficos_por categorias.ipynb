```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df_tmdb = pd.read_csv(r"C:\Users\Jp Lima\Desktop\alura_course\datasets\tmdb_5000_movies.csv")
df_tmdb.head()
```

```python
count_language = df_tmdb['original_language'].value_counts().to_frame().reset_index()
count_language.columns = ['original_language', 'total']
count_language.head()
```

```python
plt.figure(figsize=(9,6))
plt.xticks(rotation=45)
sns.barplot(data=count_language, x='original_language', y='total')
```

```python
#evitar usar 
count_language.plot(kind="pie", y="total", labels=count_language['original_language'])
```

```python
total_per_language = df_tmdb["original_language"].value_counts()
grand_total = total_per_language.sum()
total_en = total_per_language.loc["en"]
total_the_rest = grand_total - total_en
```

```python
data = {

    "lingua": ["inglês", "outros"],
    "total": [total_en, total_the_rest]
}

dados = pd.DataFrame(data)
dados
```

```python
sns.barplot(x="lingua", y="total", data=dados)
```

```python
dados.plot(kind="pie", x="lingua", y="total", labels = dados)
```
